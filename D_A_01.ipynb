```python
#Filling null  data's with Mean method
df3=data.fillna('mean')
df3.isnull().sum()
```

```python
#Listing the columns which are object but the actual values are numeric
obj_to_int=['normalized-losses', 'engine-size', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price' ]

#converting them to float64 for better pre-processing
for col in obj_to_int:
    df[col]=df[col].astype('float64')

df.dtypes
```
